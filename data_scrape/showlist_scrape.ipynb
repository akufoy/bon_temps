```python
from selenium import webdriver
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("http://showlistaustin.com/")
```

```python
events = driver.find_elements_by_class_name("h4")
```

```python
dates = driver.find_element_by_class_name("content").find_elements_by_css_selector("b")
```

```python
len(dates)
```

```python
# need to keep entries that begin with day of the week
for date in dates:
    print(date.text)
```

```python
# event text comes back with "\n" delimiting entries
events_list = []
len_list = len(events)
for event in events:
    events_list.append(event.text)
#     print(event.text)
```

```python
# splitlines() splits text on "\n"
events = []
for event in events_list:
    events.append(event.splitlines())
```
